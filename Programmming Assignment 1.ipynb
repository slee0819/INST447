### Research Question:

Do people order different foods at different times of day? For example, do people order more junk food relative to other foods late at night?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset. 

ft = pd.read_csv('InstacartOrdersByDepartment.csv')
```

```python
# calculate the total number of orders for each hour

total_orders = ft.groupby('order_hour_of_day')['num_orders_hour'].sum()
```

```python
# calculate the percentage of orders for each department at each hour

ft['percent_orders'] = ft['num_orders_hour'] / ft['tot_orders_dept']
```

```python
# pivot the table to get the percentage of orders for each department by hour

dept_hour = ft.pivot_table(values='percent_orders', index='order_hour_of_day', columns='department')
```

```python
# plot the bar chart

dept_hour.plot(kind='bar', stacked=True, figsize=(12,8))

# set the axis labels and title
plt.xlabel('Hour of the Day')
plt.ylabel('Percentage of Orders')
plt.title('Percentage of Orders by Department and Hour of the Day')


# show the legend

plt.legend(title='Department', bbox_to_anchor=(1.05, 1), loc='upper left')


# show the plot
plt.show()
```
